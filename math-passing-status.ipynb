```python
import pandas as pd

csv_path = '/kaggle/input/students-performance-in-exams/StudentsPerformance.csv'  
students_df = pd.read_csv(csv_path)

students_df.head()
```

```python
#criteria
def passing_status(score):
    return "Pass" if score >= 33 else "Fail"

# using 'math score' column to create the 'math passing status' column
students_df['math passing status'] = students_df['math score'].apply(passing_status)

students_df.head()
```

```python
# NOTE: if running this program on local system (not on kaggle website), only then this block will work

# update the data by creating new csv file with new column
updated_csv_path = 'StudentsPerformance_Updated.csv'  # Update the path as needed
students_df.to_csv(updated_csv_path, index=False)

print(f"Updated dataset saved to {updated_csv_path}")
```
